# 01 - Consolidating the data

On this stage we consolidates alld data archives into one file to be used by the algorithms.

```python
import pandas as pd
import numpy as np
```

First we import the file with the features labels to the dataset.

```python
df_labels = pd.read_csv("../UNSW-NB15/NUSW-NB15_features.csv")
```

```python
df_labels.head()
```

Now we import the data, and set the column labels.

```python
df1 = pd.read_csv("../UNSW-NB15/UNSW-NB15_1.csv", header = None)
df2 = pd.read_csv("../UNSW-NB15/UNSW-NB15_2.csv", header = None)
df3 = pd.read_csv("../UNSW-NB15/UNSW-NB15_3.csv", header = None)
df4 = pd.read_csv("../UNSW-NB15/UNSW-NB15_4.csv", header = None)
```

```python
df1.columns = df_labels['Name'].tolist()
df2.columns = df1.columns
df3.columns = df1.columns
df4.columns = df1.columns
```

```python
data = df1.append(df2)
data = data.append(df3)
data = data.append(df4)
data.head()
```

And finally we export the new dataframe as a CSV file

```python
data.to_csv('../Dataset/UNSW-NB15.csv')
```
